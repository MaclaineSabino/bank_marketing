#### criando uma coluna class_saldo baseado em balance onde agruparemos em classes os clientes de acordo com o valor do seu saldo

```python
def saldo(s):
    
    if s<0:
        return 0
    elif s==0:
        return 1
    elif s>0 and s<501:
        return 2
    
    elif s>500 and s<1001:
        return 3
    elif s>1000 and s<1501:
        return 4
    elif s>1500 and s<2001:
        return 5
    elif s>2000 and s<2501:
        return 6
    elif s>2500 and s<3001:
        return 7
    elif s>3000 and s<3501:
        return 8
    elif s>3500 and s<4001:
        return 9
    elif s>4000:
        return 10
    
```

```python
df['class_saldo']=df['balance'].apply(saldo)
```

```python
from sklearn.metrics import confusion_matrix

def fit_and_predict(nome,modelo,treino_dados,treino_marcacoes,test_dados,test_marcacoes):
    modelo.fit(treino_dados,treino_marcacoes)
    resultado = modelo.predict(test_dados)
    acertos = 0
    tamanho = len(test_marcacoes)
    for i in range(tamanho):
        if test_marcacoes[i]==resultado[i]:
            acertos = acertos+1
    perc = acertos*100/tamanho
    print('Acerto %s: %.2f%%'%(nome,perc))
    
    
    return perc
    
   
```

```python
from sklearn.naive_bayes import MultinomialNB
modeloMultinomial = MultinomialNB()
resultadoMultinomial = fit_and_predict("MultinomialNB",modeloMultinomial,treino_dados,treino_marcacoes,teste_dados,teste_marcacoes)
```

```python
from sklearn.ensemble import AdaBoostClassifier
modeloAdaBoost = AdaBoostClassifier()
resultadoAdaBoost = fit_and_predict("AdaBoostClassifier",modeloAdaBoost,treino_dados,treino_marcacoes,teste_dados,teste_marcacoes)
```

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import LinearSVC
modeloOneVsRest = OneVsRestClassifier(LinearSVC(random_state=0))
resultadoOneVsRest = fit_and_predict("OneVsRest",modeloOneVsRest,treino_dados,treino_marcacoes,teste_dados,teste_marcacoes)
```

```python
from sklearn.multiclass import OneVsOneClassifier
modeloOneVsOne = OneVsOneClassifier(LinearSVC(random_state=0))
resultadoOneVsOne = fit_and_predict("OneVsOne",modeloOneVsOne,treino_dados,treino_marcacoes,teste_dados,teste_marcacoes)
```

```python
from sklearn.ensemble import RandomForestClassifier
modelRandomForestClassifier= RandomForestClassifier()
resultadoRandomForestClassifier = fit_and_predict("RandomForestClassifier",modelRandomForestClassifier,treino_dados,treino_marcacoes,teste_dados,teste_marcacoes)
```

```python

from sklearn.ensemble import GradientBoostingClassifier

modelGradientBoostingClassifier= GradientBoostingClassifier(n_estimators=100, learning_rate=1.0, max_depth=1, random_state=0)

resultadoGradientBoostingClassifier=fit_and_predict("GradientBoostingClassifier",modelGradientBoostingClassifier,treino_dados,treino_marcacoes,teste_dados,teste_marcacoes)


```

```python
from sklearn.model_selection import cross_val_score
import numpy as np
```

```python

def folding(modelo,k):
    print("Executando com qtd de divisÃµes:  with k=%d" % k)
    scores = cross_val_score(modelo, teste_dados, teste_marcacoes, cv=k)
   
    
    return "%s: %.2f%%" % (modelo.__class__.__name__, (np.mean(scores)*100)),(np.mean(scores)*100)
```

```python
text_sai,foldingMultinomial = folding(MultinomialNB(),6)
print(text_sai)
```

```python
text_sai,foldingAdaBoost=folding(AdaBoostClassifier(),6)
print(text_sai)
```

```python
text_sai,foldingOneVsOne = folding(OneVsOneClassifier(LinearSVC(random_state=0)),6)
print(text_sai)
```

```python
text_sai,foldingOneVsRest = folding(OneVsRestClassifier(LinearSVC(random_state=0)),6)
print(text_sai)
```

```python
text_sai,foldingGradient = folding(GradientBoostingClassifier(n_estimators=100, learning_rate=1.0, max_depth=1, random_state=0),6)
print(text_sai)
```
